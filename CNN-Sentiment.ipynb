## Import Dependencies

```python
#import dependencies
import tweepy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

#import vader
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

#import Tweepy API authentication
from config import (consumer_key,
                    consumer_secret,
                    access_token,
                    access_token_secret)

# setup authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

## Getting all the Tweets

```python
#target search
target_user = ("BBCWorld", "CBSNews", "CNN", "FoxNews", "nytimes")

#set up a list to hold sentiments results
sentiments_list = []

#loop through target users
for target in target_user:
    oldest_tweet = None
    counter = 1
    
    #variable for holding sentiments
    compound_list = []
    positive_list = []
    negative_list = []
    neutral_list = []
    
    #loop through 5 pages of tweets (total 100 tweets)
    for x in range(5):
        
        #get all tweets from home feed
        public_tweets = api.user_timeline(target, max_id = oldest_tweet)

        #loop through all tweets
        for tweet in public_tweets:
            
            #run Vader on each
            results = analyzer.polarity_scores(tweet["text"])
            compound = results["compound"]
            pos = results["pos"]
            neg = results["neg"]
            neu = results["neu"]
            
#             compound_list.append(compound)
#             positive_list.append(pos)
#             negative_list.append(neg)
#             neutral_list.append(neu)
        
            #set up the new oldest_tweet value
            oldest_tweet = tweet["id"] - 1
        
            #Store the average sentiments
            sentiments_list.append({"User": target,
                                "Compound": compound,
                                "Positive": pos,
                                "Negative": neg,
                                "Neutral": neu,
                                "Tweets Ago": counter,
                                "Date": tweet["created_at"]})
        
            #add to counter for the count
            counter += 1
```

```python
sentiments_df = pd.DataFrame(sentiments_list)
sentiments_df.head()
```

## Sentiments Graph

```python
plt.figure(figsize=(10, 10))
BBCWorld = plt.scatter(BBCWorld_df["Tweets Ago"],
            BBCWorld_df["Compound"],
            color="blue",
            edgecolor="black",
            alpha=0.8,
            linewidths=1,
            s=150,
            label="BBCWorld")
CBSNews = plt.scatter(CBSNews_df["Tweets Ago"],
            CBSNews_df["Compound"],
            edgecolor="black",
            color="orange",
            linewidths=1,
            alpha=0.8,
            s=150,
            label="CBS News")
CNN = plt.scatter(CNN_df["Tweets Ago"],
            CNN_df["Compound"],
            edgecolor="black",
            color="green",
            linewidths=1,
            alpha=0.8,
            s=150,
            label="CNN")
FoxNews = plt.scatter(FoxNews_df["Tweets Ago"],
            FoxNews_df["Compound"],
            edgecolor="black",
            color="red",
            linewidths=1,
            alpha=0.8,
            s=150,
            label="Fox News")
NYTimes = plt.scatter(NYTimes_df["Tweets Ago"],
            NYTimes_df["Compound"],
            edgecolor="black",
            color="purple",
            linewidths=1,
            alpha=0.8,
            s=150,
            label="New York Times")

#incorporate other graph properties
now = datetime.now()
now = now.strftime("%Y-%m-%d %H:%M")
plt.title(f"Sentiment Analysis of Tweets ({now})", fontsize=20)
plt.grid()
plt.xlim(105, -5)
plt.ylabel("Tweet Polarity")
plt.xlabel("Tweets Ago")
plt.legend(title="Media Outlet", loc="best")
plt.savefig("output_analysis/sentiment_tweet_analysis.png")
plt.show()
```

## Average Sentiments Graph

```python
sentiments_group = sentiments_df.groupby(["User"])
compound_average = sentiments_group["Compound"].mean()
negative_average = sentiments_group["Negative"].mean()
positive_average = sentiments_group["Positive"].mean()
neutral_average = sentiments_group["Neutral"].mean()

sentiments_ave_df = pd.DataFrame({"Compound": compound_average,
                                  "Negative": negative_average,
                                  "Positive": positive_average,
                                  "Neutral": neutral_average})
sentiments_ave_df
```

```python
plt.figure(figsize=(10,7))
compound_average.plot(kind="bar", width=1, linewidth=1, edgecolor="black")
plt.hlines(0, -.5, 5, alpha=.25)
plt.title(f"Overall Media Sentiment ({now})", fontsize=15)
plt.xlabel("News Outlet", fontsize=12)
plt.ylabel("Average Compound Polarity", fontsize=12)
plt.tick_params(rotation=0)
plt.xlim(-.5,4.5)
plt.savefig("output_analysis/overall_sentiment.png")
```
